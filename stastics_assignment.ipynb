```python
#let's run the code to see the SPPS outputs and check how much SPSS is correct with figures , here output has been already obtaines by SPSS

import statsmodels.api as sm
import pandas as pd

# Replace SPSS missing codes
df = df.replace([-8, -9], pd.NA)

# Define dependent variable
y = df["buyingpreference"]

# Define predictors (exactly as SPSS used)
X = df[
    ["black","green","brown",
     "Glas","BiobasedMaterials","RecycledPlastics",
     "VeganBlume","CrueltyFree","QRCode","Refill","Price"]
]

# Add constant (intercept)
X = sm.add_constant(X)

# Run regression
model = sm.OLS(y, X, missing="drop").fit()

# Show SPSS-style output
print(model.summary())
```

```python
#now let's run the origiinal Python wise and see about the regressions and poutput

import statsmodels.api as sm

# Ensure X has a constant added
X = sm.add_constant(X, prepend=False) # Use prepend=False to add constant at the end if it's already there

# Run the regression, dropping observations with missing values
model = sm.OLS(y, X, missing="drop").fit()

model.summary()
```

# Let's make the charts now to understand , which pacakging feature are more effecient , postive and negartive effects and which attributes domiantes

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

df = df.replace([-8, -9], pd.NA)

y = df["buyingpreference"]
X = df[["black","green","brown","Glas","BiobasedMaterials","RecycledPlastics",
        "VeganBlume","CrueltyFree","QRCode","Refill","Price"]]
X = sm.add_constant(X)

model = sm.OLS(y, X, missing="drop").fit()

utilities = model.params
utilities_no_const = utilities.drop("const")
importance = (np.abs(utilities_no_const) / np.abs(utilities_no_const).sum()) * 100
```

# part_worth_utilities.png and attribute_importance.png

```python
import matplotlib.pyplot as plt

# Chart A: Part-worth utilities
plt.figure()
utilities_no_const.sort_values().plot(kind="barh")
plt.title("Part-worth Utilities of Packaging Attributes")
plt.xlabel("Utility")
plt.tight_layout()
plt.savefig("part_worth_utilities.png")
plt.show()

# Chart B: Attribute importance
plt.figure()
importance.sort_values().plot(kind="barh")
plt.title("Relative Importance of Packaging Attributes (%)")
plt.xlabel("Importance (%)")
plt.tight_layout()
plt.savefig("attribute_importance.png")
plt.show()
```

```python
# Predicted values
df_clean = df.replace([-8, -9], pd.NA)
df_clean["predicted"] = model.predict(X)

import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df_clean["predicted"], df_clean["buyingpreference"], alpha=0.5)
plt.xlabel("Predicted Buying Preference")
plt.ylabel("Observed Buying Preference")
plt.title("Predicted vs Observed Buying Preference")
plt.tight_layout()
plt.savefig("regression_fit.png")
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt

# Load
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Main stastics ASSIGNMENT /Dataset data View.xlsx")
df = df.replace([-8, -9], np.nan)

# Model
y = df["buyingpreference"]
X = df[["black","green","brown","Glas","BiobasedMaterials","RecycledPlastics",
        "VeganBlume","CrueltyFree","QRCode","Refill","Price"]]
X = sm.add_constant(X)
model = sm.OLS(y, X, missing="drop").fit()

# Part-worth utilities
utilities = model.params.drop("const")

plt.figure(figsize=(8,6))
utilities.sort_values().plot(kind="barh")
plt.title("Part-worth Utilities of Packaging Attributes")
plt.xlabel("Utility")
plt.tight_layout()
plt.show()

# Attribute importance
importance = abs(utilities) / abs(utilities).sum() * 100

plt.figure(figsize=(8,6))
importance.sort_values().plot(kind="barh")
plt.title("Relative Importance of Attributes (%)")
plt.xlabel("Importance (%)")
plt.tight_layout()
plt.show()
```
